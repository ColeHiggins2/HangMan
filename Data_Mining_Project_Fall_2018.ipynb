```python
# Regression Model
# IMPORTANT NOTE: Please reimport datasets before running each model. Not doing so 
# will result in an error due to the way the data gets modified before each model.

from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Initialize model
regr = linear_model.LinearRegression()

# Split training data
_ = train.pop('Date')
y = train.pop('Sales')
X = train

# Divide data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= .2, random_state= 40)

# Fit data
regr.fit(X_train, y_train)

# Predict values
prediction = regr.predict(X_test)

# Evaluate model

print('Coefficients: \n', regr.coef_)

print('Coefficient of multiple determination: %.2f' % r2_score(y_test, prediction))

plt.scatter(y_test, prediction)
plt.show()
```

```python
# Clustering 
# IMPORTANT NOTE: Please reimport datasets before running each model. Not doing so 
# will result in an error due to the way the data gets modified before each model.

from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans

_ = train.pop('Date')
y = train.pop('Sales')
X = train

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= .2, random_state= 40)

ks = range(1, 10)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(X_train)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()


```

```python
#AA 
# IMPORTANT NOTE: Please reimport datasets before running each model. Not doing so 
# will result in an error due to the way the data gets modified before each model.

#adding extra columns to dataframe for aprioi algo
store.insert(10, "DistanceBin", 1115) 
store.insert(11, "SalesBin", 1115) 
store.insert(12, "Assortment2", 1115)

#disbale warning
pd.options.mode.chained_assignment = None  # default='warn'

#creates list of first 21 sales to plug into AA model
ind_mean_sales = [0] * 1115
for k in range (0, 20):
    for i in range (0, 1114):
        ind_mean_sales[i] += train.Sales[i]

#takes average of each store sales 
for i in range (0, 1114):
    ind_mean_sales[i] += (ind_mean_sales[i] / 21)

#poplating assortment2 column
for i in range (0,1115):     
    if store.Assortment[i] == 'a':
        store.Assortment2[i] = 'AssortmentA'
        
    elif store.Assortment[i] == 'b':
        store.Assortment2[i] = 'AssortmentB'
    
    elif store.Assortment[i] == 'c':
        store.Assortment2[i] = 'AssortmentC' 
    else:
        store.Assortment2[i] = 'AssortmentD'
        
#binning values for sales
for i in range (0,1115):     
    if ind_mean_sales[i] < 100000:
        store.SalesBin[i] = 'LowProfit'
        
    elif ind_mean_sales[i]  < 300000:
        store.SalesBin[i] = 'MediumProfit'
       
    else:
        store.SalesBin[i] = 'HighProfit'

#binning values for sales
for i in range (0,1115):     
    if store.CompetitionDistance[i] < 3850:
        store.DistanceBin[i] = 'LowCompDistance'
        
    elif store.CompetitionDistance[i] < 10600:
        store.DistanceBin[i] = 'MediumCompDistance'
       
    else:
        store.DistanceBin[i] = 'HighCompDistance'

    
#turns columns intoo lists 
store_type = store.StoreType.tolist()
store_assortment = store.Assortment2.tolist()
store_distance = store.DistanceBin.tolist()
store_sales = store.SalesBin.tolist()

#combines sep lists into lists of nested lists for aprori algo
records = list(it.chain.from_iterable(it.zip_longest(store_type,store_assortment, store_distance,store_sales)))

#nests the tuple into list of nested list
i=0
new_list=[]
while i<len(records):
    new_list.append(records[i:i+4])
    i+=4

#populating aapriori algo  
association_rules = apriori(new_list, min_support = 0.3, min_confidence = 0.3, min_lift = 1)
association_results = list(association_rules)
```

```python
#Prints ternary associations
#NOTE single chars = STORE TYPE
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    if (len(items) != 3):
        continue
    print("Rule: " + items[0] + " -> " + items[1] + " -> " + items[2])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

<h3>Association Rule Analysis</h3>
<p>The bulk of the work that had to be done to prepare the data for the association rule analysis was to bin the numerical data. All of the rules we found basically imply a 10% to 16% increase in the likelihood of those events occuring together versus them occuring on their own. The first rule states that a low distance of competition implies a store carrying AssortmentA which implies a Medium Profit. The second rule states that store type a implies carrying AssortmentA which also implies Medium Profit.
